```python
# if it is a model with one layer and one nueron 
model = nn.Sequential(nn.Linear(1, 1))
```

```python
#   losss and optimizer, here losss is criterion because loss is is not callable since it is a tensor object
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

```python
xs = torch.tensor([[-1.0], [0.0], [1.0], [2.0], [3.0], [4.0]], dtype=torch.float32)

ys = torch.tensor([[-3.0], [-1.0], [1.0], [3.0], [5.0], [7.0]],dtype=torch.float32)
```

```python
# Data
# Training features ['Years of experience', 'performance score', 'projects completed']
Features = torch.tensor([[1, 60, 2], [2, 65, 3], [3, 70, 4], [4, 72, 5], [5,  78,  6], [6,  82,  7], [7,  85,  8], [8,  88,  9], [9,  92, 10], [10, 95, 12]], dtype=torch.float32)

Salary = torch.tensor([[500], [700], [900], [1100], [1400], [1700], [2000], [2300], [2700], [3200]], dtype=torch.float32)
```

```python
# TRianing
for _ in range(500):
    optimizer.zero_grad()
    output = model(xs)
    loss = criterion(output, ys)
    loss.backward()
    optimizer.step()
    print("Loss = ",loss)
```

```python
# Predict
with torch.no_grad():
    print(model(torch.tensor([[10.0]], dtype=torch.float32)))
```
